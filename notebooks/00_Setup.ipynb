```python
spark.sql("CREATE DATABASE IF NOT EXISTS dbestudiantes")
spark.sql("USE dbestudiantes")
```

```python
spark.sql("""
          create table if not exists meta_jobs(
              job_name string,
              layer string,
              start_time timestamp,
              end_time timestamp,
              rows long,
              id_ingesta_last int,
              status string,
              message string
          )
          using delta
          """)
```

```python
spark.sql("""
          create table if not exists logs_job(
              log_time timestamp,
              job_name string,
              layer string,
              level string,  -- INF,WARN,ERROR
              message string
          )
          using delta
             
          """)
```
